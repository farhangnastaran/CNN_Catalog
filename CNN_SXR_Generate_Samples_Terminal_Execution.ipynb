```python
import subprocess
import time

# --- Execution Timer Start ---
start_time = time.time()

# Paths to Python interpreter and target scripts
python_exe = r"directory to the py_env\Scripts\python.exe"
script_path_1 = r"directory to files\CNN_SXR_Generate_Samples_Train_Manual_Set.py"
script_path_2 = r"directory to files\CNN_SXR_Generate_Samples_Prediction.py"

# --- Run First Script (Training Sample Generation) ---
result = subprocess.run(
    [python_exe, script_path_1],
    capture_output=True,
    text=True
)

# if it failed (returncode != 0), raise an error
if result.returncode != 0:
    # print stdout and stderr
    print("STDOUT:")
    print(result.stdout)
    print("\nSTDERR:")
    print(result.stderr)
    print(f"\nProcess exited with code {result.returncode}")
    raise RuntimeError("Subprocess failed. See errors above.")

# --- Run Second Script (Prediction Sample Generation) ---
result_2 = subprocess.run(
    [python_exe, script_path_2],
    capture_output=True,
    text=True
)

# if it failed (returncode != 0), raise an error
if result_2.returncode != 0:
    # print stdout and stderr
    print("STDOUT:")
    print(result_2.stdout)
    print("\nSTDERR:")
    print(result_2.stderr)
    print(f"\nProcess exited with code {result_2.returncode}")
    raise RuntimeError("Subprocess failed. See errors above.")

# --- Execution Summary ---
end_time = time.time()
elapsed = end_time - start_time

print(f"Execution time: {elapsed:.2f} seconds ({elapsed/60:.2f} minutes)")
```
